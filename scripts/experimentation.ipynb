```python
from GoogleNews import GoogleNews
from bs4 import BeautifulSoup
import pandas as pd

def scrape_google_news(keyword, num_pages=5):
    googlenews = GoogleNews(lang='en', period='7d')
    googlenews.search(keyword)

    news_list = []

    for i in range(1, num_pages + 1):
        googlenews.getpage(i)
        result = googlenews.result()

        for news in result:
            title = news['title']
            media = news['media']
            date = news['date']
            desc = news['desc']
            link = news['link']

            news_list.append({
                'Title': title,
                'Media': media,
                'Date': date,
                'Description': desc,
                'Link': link
            })

    return news_list

keyword = 'cryptocurrency'
news_data = scrape_google_news(keyword, num_pages=2)
df = pd.DataFrame(news_data)
df
```
